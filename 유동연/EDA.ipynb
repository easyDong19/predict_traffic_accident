```python
import os
import random
import numpy as np

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(42)
```

```python

attacker_age = pd.DataFrame(train_org['가해운전자 연령'])
attacker_age = attacker_age[attacker_age['가해운전자 연령'] != '미분류']
attacker_age['가해운전자 연령'] = attacker_age['가해운전자 연령'].str.strip()
attacker_age['가해운전자 연령'] = attacker_age['가해운전자 연령'].replace(regex=r'[/세/이상]', value='').astype('int64')
attacker_age['가해운전자 연령']

attacker_age['Age_cut'] = pd.cut(attacker_age['가해운전자 연령'], bins = [0,9,19,29,39,49,59,100], labels=['10대미만','10대','20대','30대','40대','50대','60대이상'])
sns.set(rc = {'figure.figsize' : (10,8), 'font.family' : 'AppleGothic'})
sns.distplot(attacker_age['가해운전자 연령'] , rug=True, hist=True, kde=True)

plt.show()
```

```python
rc('font', family='AppleGothic') 			## 이 두 줄을 
plt.rcParams['axes.unicode_minus'] = False  ## 추가해줍니다. 

attacker_age['Age_cut'].value_counts().plot.bar()
```
